```python
# neighborhood market tracker
nmt_data = '/kaggle/input/redfin-housing-market-data/neighborhood_market_tracker.tsv000'
nmt_df = pd.read_csv(nmt_data, sep='\t')
```

```python
nmt_df['year'] = nmt_df['period_begin'].astype(str).str[:4]
nmt_clean_df = nmt_df[['city', 'state', 'state_code', 'parent_metro_region','year', 'property_type', 'median_sale_price', 'median_list_price',  'homes_sold',   'new_listings',  'inventory', 'avg_sale_to_list', 'sold_above_list']]
```

```python
nmt_clean_df['city_state'] = nmt_clean_df['city'] + ', ' + nmt_clean_df['state_code']
```

```python
metro_city_df = nmt_clean_df.copy()
```

```python
metro_city_df[metro_city_df['parent_metro_region'] == metro_city_df['city_state']]
```

```python
metro_city_df = metro_city_df[['city', 'state', 'year', 'property_type', 'median_sale_price', 'median_list_price',  'homes_sold',   'new_listings',  'inventory', 'avg_sale_to_list', 'sold_above_list']]
final_metro_city_df = metro_city_df.sort_values(by=['state','city','year'])
final_metro_city_df
```
