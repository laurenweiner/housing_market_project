```python
import pandas as pd
import time
```

```python
redfin_data = pd.read_csv('county_df.csv')
census_data = pd.read_csv('co_est2020.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
redfin_data['year'] = redfin_data['period_begin'].astype(str).str[:4]
redfin_clean_df = redfin_data[['region', 'state', 'state_code', 'parent_metro_region','year', 'property_type', 'median_sale_price', 'median_list_price',  'homes_sold',   'new_listings',  'inventory', 'avg_sale_to_list', 'sold_above_list']]
redfin_clean_df['population'] = ''
redfin_clean_df.head()
```

```python
avg_redfin_county = redfin_clean_df.groupby(['state','region','year']).mean()
avg_redfin_county
```

```python
final_redfin_county = avg_redfin_county.copy()
final_redfin_county = final_redfin_county.reset_index()
final_redfin_county['population']=''
final_redfin_county.head()
```

```python
for i in range(0,len(final_redfin_county)):
    r = final_redfin_county.region[i]
    p = final_redfin_county.population[i]
    j = census_data.loc[census_data['countystate'] == r]
    k = pd.DataFrame(j)
    l = k.reset_index()
    
    try:
        m = l.iloc[0,13] #2012
        n = l.iloc[0,14] #2013
        o = l.iloc[0,15] #2014
        q = l.iloc[0,16] #2015
        s = l.iloc[0,17] #2016
        t = l.iloc[0,18] #2017
        u = l.iloc[0,19] #2018
        v = l.iloc[0,20] #2019
        w = l.iloc[0,22] #2020
        time.sleep(1)

        if final_redfin_county.year[i] == '2012':
            final_redfin_county.population[i] = m

        elif final_redfin_county.year[i] == '2013':
            final_redfin_county.population[i] = n

        elif final_redfin_county.year[i] == '2014':
            final_redfin_county.population[i] = o

        elif final_redfin_county.year[i] == '2015':
            final_redfin_county.population[i] = q

        elif final_redfin_county.year[i] == '2016':
            final_redfin_county.population[i] = s

        elif final_redfin_county.year[i] == '2017':
            final_redfin_county.population[i] = t

        elif final_redfin_county.year[i] == '2018':
            final_redfin_county.population[i] = u

        elif final_redfin_county.year[i] == '2019':
            final_redfin_county.population[i] = v

        elif final_redfin_county.year[i] == '2020':
            final_redfin_county.population[i] = w

        else:
            final_redfin_county.population[i] = ''
            
    except:
        pass
```
